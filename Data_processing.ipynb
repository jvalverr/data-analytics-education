# Advanced Large Language Models & Visualization Tools for Data Analytics Learning

We present the data analysis and processing for the research project *Advanced Large Language Models & Visualization Tools for Data Analytics Learning*. 

We conducted a case study with a cohort of 59 participants affiliated with higher education programs at Tecnologico de Monterrey, a private university in Mexico. The cohort included 43 undergraduate students enrolled in a Data Analytics course and 16 professionals from a continuing education course focused on Data Analytics. As part of the case study, participants developed a data analytics project, and later answers a survey about their experience.

The data is organized into several columns, including:

- Gender
- Age range
- Current role (Professional or Student)
- Campus
- Affiliation
- Programming experience
- Data analytics experience
- Experience using tools like Python, Colab, etc.
- Experience using ChatGPT in general
- Experience using ChatGPT for programming
- Experience using ChatGPT for data analytics
- Experience using generative APIs
- Time required to finish activity 1
- Time required to finish activity 2
- Time required to finish activity 3
- Ease of use
- Speed of result
- Appropriateness
- Helpful
- Correctness

Download the data from [the paper's supplementary material](https://www.frontiersin.org/journals/education/articles/10.3389/feduc.2024.1418006/abstract#supplementary-material)

```python
def get_uniques(dataframe: pd.DataFrame) -> dict:
  """
  Gets the unique values for each column of a dataframe

    Parameters
    - dataframe : DataFrame
        The dataframe to be processed

    Returns:
    - unique_values : dict
        A dictionary where the keys are the colums of the dataframe, and the values are the unique values of the column
  """
  unique_values = {}

  for column in dataframe:
    unique_values[column] = pd.unique(dataframe[column])

  return unique_values
```

```python
def pie_chart(data, column, colors=None, title=True, subpath="", article_image=False):
    """
    Generate a pie chart based on the given data and column.

    Parameters:
    - data: DataFrame
        The input data containing the column to be used for the pie chart.
    - column: str
        The column name to be used for the pie chart.
    - colors: dict, optional
        A dictionary mapping values in the column to colors for the pie chart.
    - title: bool, optional
        Whether to display a title for the pie chart. Default is True.
    - subpath: str, optional
        A subpath to be included in the saved image file name. Default is an empty string.
    - article_image: bool, optional
        Whether the image is intended for an article. Default is False.

    Returns:
    None
    """
    df_pie = data[column].value_counts()
    plot = df_pie.plot.pie(title=f"Number of participants by {column.lower()}" if title else "", 
                           startangle=90, 
                           autopct='%1.0f%%', 
                           ylabel="", 
                           legend=None,
                           colors=[colors[v] for v in df_pie.keys()] if colors is not None else None)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    path = "Pie" if not article_image else "Journal"
    subpath = f" for {subpath.lower()}" if subpath else ""
    plot.get_figure().savefig(f"images/{path}/Pie - Number of participants by {column.lower()}{subpath}.eps", bbox_inches='tight', dpi=300, format='eps')
```

```python
def bar_chart(data, column, title=True, subpath="", article_image=False):
    """
    Generate a bar chart based on the given data and column.

    Parameters:
    - data: DataFrame
        The input data containing the participants' information.
    - column: str
        The column name in the data to be used for the bar chart.
    - title: bool, optional
        Whether to display the title of the chart. Default is True.
    - subpath: str, optional
        Additional information to be included in the chart's filename. Default is an empty string.
    - article_image: bool, optional
        Whether the chart is intended for an article. Default is False.

    Returns:
    None
    """
    ax = sns.countplot(x=column, data=data, order=data[column].value_counts().index, color='#d5a8d5')
    ax.set_title(f"Number of participants by {column.lower()}" if title else "")
    ax.set_ylabel("Number of participants")
    ax.set_yticks(range(0, 30, 5))
    ax.set_xlabel(None)
    plt.xticks(rotation=90)
    plt.show()
    path = "Bar" if not article_image else "Journal"
    subpath = f" for {subpath.lower()}" if subpath else ""
    ax.figure.savefig(f"images/{path}/Bar - Number of participants by {column.lower()}{subpath}.eps", bbox_inches='tight', dpi=300, format='eps')
```

```python
def bar_chart_hue(data, column, hue, title=True, subpath="", article_image=False):
    """
    Generate a bar chart with hue using the given data.

    Parameters:
    - data: DataFrame
        The input data containing the required columns.
    - column: str
        The column name to be plotted on the x-axis.
    - hue: str
        The column name to be used for grouping and coloring the bars.
    - title: bool, optional
        Whether to display the chart title. Default is True.
    - subpath: str, optional
        Additional subpath to be added to the image file name. Default is an empty string.
    - article_image: bool, optional
        Whether the image is intended for an article. Default is False.

    Returns:
    None
    """
    ax = sns.countplot(x=column, data=data, hue=hue, order=data[column].value_counts().index)
    ax.set_title(f"Number of participants by {column.lower()} and {hue.lower()}" if title else "")
    ax.set_ylabel("Number of participants")
    ax.set_yticks(range(0, 30, 5))
    ax.set_xlabel(None)
    plt.xticks(rotation=90)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    path = "Bar_hue" if not article_image else "Journal"
    subpath = f" for {subpath.lower()}" if subpath else ""
    ax.figure.savefig(f"images/{path}/Bar - Number of participants by {column.lower()} and {hue.lower()}{subpath}.eps", bbox_inches='tight', dpi=300, format='eps')
```

### Lollipop charts

```python
def metrics_chart(df, column, column_colors, excluded, metrics, options, title=True, article_image=False):
    """
    Generate a metrics chart based on the given data.

    Parameters:
    - df (pandas.DataFrame): The input dataframe containing the data.
    - column (str): The column name in the dataframe to group the data by.
    - column_colors (list): A list of colors for each unique value in the 'column' column.
    - excluded (list): A list of values to exclude from the chart.
    - metrics (list): A list of metrics to display on the chart.
    - options (list): A list of options for each metric.
    - title (bool, optional): Whether to display the title on the chart. Defaults to True.
    - article_image (bool, optional): Whether to save the chart as an image for an article. Defaults to False.

    Returns:
    None
    """

    # Create an empty dictionary to store the metrics data for each unique value in the 'column' column
    df_metrics = {}

    # Iterate over each unique value in the 'column' column
    for value in df[column].unique():
        # Filter the dataframe based on the current value
        df_temp = df[df[column]==value].melt(value_vars=metrics, var_name="Metric", value_name="Answer").value_counts().reset_index().rename(columns={"count": "People"})
        
        # Check if any metrics or options are missing for the current value
        for m in metrics:
            for o in options:
                if len(df_temp[(df_temp["Metric"] == m) & (df_temp["Answer"] == o)]) == 0:
                    # Add missing metrics and options with 0 count
                    df_temp = df_temp._append({"Metric": m, "Answer": o, "People": 0}, ignore_index=True)
        
        # Set the order of metrics and options for sorting
        df_temp["Metric"] = pd.Categorical(df_temp["Metric"], ordered=True, categories=metrics)
        df_temp["Answer"] = pd.Categorical(df_temp["Answer"], ordered=True, categories=options)
        
        # Sort the dataframe by metric and answer
        df_temp = df_temp.sort_values(by=["Metric", "Answer"], ascending=False)
        
        # Store the metrics data for the current value in the dictionary
        df_metrics[value] = df_temp

    # Define the colors for each metric
    colors = {val: sns.color_palette("Set2").as_hex()[pos] for pos, val in enumerate(df_metrics[df[column][0]]["Metric"].unique())}

    # Create the figure and set the size
    fig = plt.figure(figsize=(14, 10))

    # Define the range for the y-axis
    plt_range = range(1, len(df_metrics[df[column][0]]) + 1)

    # Plot the horizontal lines for each metric
    plt.hlines(y=plt_range, xmin=df_metrics[df[column].unique()[0]]["People"], xmax=df_metrics[df[column].unique()[1]]["People"], color=df_metrics[df[column][0]]["Metric"].apply(lambda x: colors[x]), linewidth=3)

    # Plot the horizontal lines to separate the metrics
    for i in range(1, len(metrics)):
        plt.axhline(y=i * len(metrics) + 0.5, color='black', linestyle='--', alpha=0.3, linewidth=3)

    # Plot the scatter points for each value in the 'column' column
    for pos, val in enumerate(df[column].unique()):
        if val not in excluded:
            plt.scatter(df_metrics[val]["People"], plt_range, label=val, color=column_colors[pos], zorder=3)

    # Define the y-axis labels based on the answer values in the dataframe
    yticks = {'Traditional method': 'Approach 1', 'ChatGPT as an external assistant': 'Approach 2', 'LIDA with GPT integration': 'Approach 3', 'None':'None'}
    new_yticks = [yticks[answer] for answer in df_metrics[df[column].unique()[0]]["Answer"] ]

    # Set the y-axis labels
    plt.yticks(plt_range, new_yticks)
    plt.rc('xtick', labelsize=15)
    plt.rc('ytick', labelsize=15)
    
    # Set the x-axis label
    plt.xlabel("Number of participants")
    
    # Create legend lines for each metric
    lines = [Line2D([0], [0], color=colors[x], label = x, linewidth=3) for x in metrics]
    handles, labels = plt.gca().get_legend_handles_labels()
    
    # Add the legend to the plot
    plt.legend(lines + handles, metrics + labels, ncol=1, bbox_to_anchor=(1.01, 0.5), loc="center left", frameon=False, fontsize=15)
    
    # Adjust the layout and display the plot
    plt.tight_layout()
    plt.show()
    
    # Save the plot as an image if article_image is True
    path = "Lollipop" if not article_image else "Journal"
    fig.savefig(f"images/{path}/Lollipop - Metrics evaluation by {column.lower()}.eps", bbox_inches='tight', dpi=300, format='eps')
```

## Stacked plots

### Time to finish activities

```python
# Filtering the data for the time required to finish each activity

df_time = df[['Current role', 'Time required to finish activity 1', 'Time required to finish activity 2', 'Time required to finish activity 3']]
df_time.head()
```

```python
# Grouping the data by the current role and the time required to finish each activity
# The values are normalized to show the percentage of each category

df_time_1 = df_time[['Current role', 'Time required to finish activity 1']].groupby(['Current role']).value_counts(normalize=True).reset_index(name='Count 1')
df_time_2 = df_time[['Current role', 'Time required to finish activity 2']].groupby(['Current role']).value_counts(normalize=True).reset_index(name='Count 2')
df_time_3 = df_time[['Current role', 'Time required to finish activity 3']].groupby(['Current role']).value_counts(normalize=True).reset_index(name='Count 3')

# Multiplying the values by 100 to get the percentage

df_time_1['Count 1'] = df_time_1['Count 1'] * 100
df_time_2['Count 2'] = df_time_2['Count 2'] * 100
df_time_3['Count 3'] = df_time_3['Count 3'] * 100
```

```python
# Pivoting the data to create a table with the percentage of each category for each current role
# The columns are the time required to finish each activity, and the rows are the current roles. The values are the percentage of each category

df_time_1_pt = df_time_1.pivot_table(index='Current role', columns='Time required to finish activity 1', values='Count 1').reset_index()
df_time_2_pt = df_time_2.pivot_table(index='Current role', columns='Time required to finish activity 2', values='Count 2').reset_index()
df_time_3_pt = df_time_3.pivot_table(index='Current role', columns='Time required to finish activity 3', values='Count 3').reset_index()

# The 'Less than 5 minutes' category is added to the first activity, as it was not present in the original data

df_time_1_pt['Less than 5 minutes'] = 0

# Reordering the columns

df_time_1_pt = df_time_1_pt[['Current role','Less than 5 minutes','5 to 10 minutes', '10 to 15 minutes','15 to 30 minutes', 'More than 30 minutes']]
df_time_2_pt = df_time_2_pt[['Current role','Less than 5 minutes','5 to 10 minutes', '10 to 15 minutes','15 to 30 minutes', 'More than 30 minutes']]
df_time_3_pt = df_time_3_pt[['Current role','Less than 5 minutes','5 to 10 minutes', '10 to 15 minutes','15 to 30 minutes', 'More than 30 minutes']]

# Changing the current role names to include the method used to finish the activities

df_time_1_pt['Current role'] = df_time_1_pt['Current role'].map(lambda x: str(x) + " by using traditional approach")
df_time_2_pt['Current role'] = df_time_2_pt['Current role'].map(lambda x: str(x) + " by using ChatGPT")
df_time_3_pt['Current role'] = df_time_3_pt['Current role'].map(lambda x: str(x) + " by using LIDA+GPT")

display(df_time_1_pt)
display(df_time_2_pt)
display(df_time_3_pt)
```

```python
# Concatenate the dataframes df_time_1_pt, df_time_2_pt, and df_time_3_pt into a single dataframe
frames = [df_time_1_pt, df_time_2_pt, df_time_3_pt]
df_concatenated = pd.concat(frames).set_index('Current role')

# Create a figure and axis object
plt.figure(figsize=(10,6))
p = df_concatenated.plot(kind='barh', stacked=True)

# Set the title, legend, and label for the plot
plt.title("Time Required to Finish Activities")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.ylabel("")
plt.xlabel('Percentage of Participants')

# Display the plot
plt.show()

# Save the plot as an EPS file
p.figure.savefig("images/Journal/Bar stacked - Time Required to Finish Activities.eps", bbox_inches='tight', dpi=300, format='eps')
```
